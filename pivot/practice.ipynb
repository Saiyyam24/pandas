## Pivot Table


import pandas as pd

```python
import pandas as pd
```

```python
df = pd.DataFrame(
    {
        "date": ["5/1/2017","5/2/2017","5/3/2017","5/1/2017", "5/2/2017","5/3/2017","5/1/2017", "5/2/2017","5/3/2017"],
        "city":["new york","new york","new york","mumbai","mumbai","mumbai","beijing","beijing","beijing"],
        "temprature": [65, 66, 68, 75,78,82,80,77,78],
        "humidity": [56, 58, 60, 80, 83,85,26,30,35],
    }
)
df
```

```python
df.to_csv('weather.csv')
```

```python
df.pivot(index="date",columns="city")
```

```python
df = pd.read_csv("weather2.csv")
df
```

```python
df.pivot_table(index="city",columns="date")
```

```python
# margins help in aggregate all the according to val
df.pivot_table(index="city",columns="date",margins=True)
```

```python
df2 = pd.read_csv("weather3.csv")
df2['date']=pd.to_datetime(df2['date'])
```

```python
df2
```

```python
type(df2['date'][0])
```

```python
# pivot_table grouper help in analysis of particular month

df2.pivot_table(index = pd.Grouper(freq='M',key='date'),columns='city')
```
