```python
import pandas as pd
```

```python
df = pd.read_csv("weather_data (1).csv",parse_dates=["day"])
df
```

```python
type(df.day[0])
```

```python
df.set_index('day',inplace=True)
df
```

```python
new_df = df.fillna(0)
new_df
# here we basically assign the na value 0
```

```python
# here we assign na value and specify according to each columns
new_df = df.fillna({
    'temperature' : 0,
    'windspeed' : 0,
    'event' : "noevent"

})
new_df
```

```python
new_df = df.fillna(method = "ffill")
# it basically carry forward the previous year value
new_df
# bfill use for next year value
```

```python
new_df=df.fillna(method="ffill",limit=1)
new_df
```

```python
# linear middle value present
new_df=df.interpolate(method='time')
new_df
```

```python
# remove columns which have null value
new_Df = df.dropna()
new_Df
```

```python
# remove null values where all columns are null
new_Df = df.dropna(how="all")
new_Df
```

```python
df
```

```python
# when we have minnimum 1 value present in a row thresh present
new_df = df.dropna(thresh=1)
new_df
```

```python
dt = pd.date_range("1-1-2017","1-11-2017")
idx = pd.DatetimeIndex(dt)
df = df.reindex(idx)
df = df.interpolate()
df=df.fillna(method = "ffill")
df
```
