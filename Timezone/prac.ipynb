```python
import pandas as pd 

df = pd.read_csv('fb.csv',parse_dates=["Date"],index_col="Date")
df
```

```python
df.index
```

```python
df.shift(1)
```

```python
df['prev day price'] = df['Price'].shift(1)
df
```

```python
df['1 day price'] = df['Price']-df['prev day price']
df
```

```python
df['prev 5 day price'] = df['Price'].shift(5)
df['5 day % return'] = ((df['Price']-df['Price'].shift(5))*100)/df['Price'].shift(5)
df
```

```python
# shift dates
df.columns = df.columns.str.strip()  # Removes leading/trailing spaces
```

```python
df
```

```python
df = df.to_frame(name='Price')
df
```
